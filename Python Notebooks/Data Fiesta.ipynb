##**Calculating no. of customers in each region**

##

```python


customer_region_data=pd.read_csv('Data/CUSTOMERS.csv')

df1=pd.DataFrame()

df1=customer_region_data.groupby('Region Code').count()

#df1.head(30)

df2=pd.DataFrame()

#regcode=df1['Region Code']

count=df1['Division']

#df2['Region Code']=regcode

region_name=['Canada','Western','Southern','Northeast','Central','International']

#df2['Region Name']=region_name

df2['Region Count']=count

df2.head(20)





```

Total customers

```python
total_customers=df2['Region Count'].sum()
print(f"Total customers={total_customers}")
```

#Plotting no.of customers(Region wise) in a pie chart

```python
import matplotlib.pyplot as plt



plot = df2.plot.pie(subplots=True, figsize=(9, 9),autopct='%1.1f%%')

plt.legend(region_name,loc="lower right")


```

From the above pie chart,we can clearly see that region 5(International) has the highest no.of customers

#**Calculating budget monthwise and yearwise**

```python
budget=pd.read_excel('Data/MERGERALL.xlsx')
```

```python


#budget_data=pd.DataFrame()

budget_yearwise_count=budget.groupby('BudgetYear').sum()

budget_yearwise_count.head(30)
```

```python
years=budget['BudgetYear'].unique()

budget_amount=budget_yearwise_count['Budget Amount']

budget_year=pd.DataFrame()

budget_year['Budget Amount']=budget_yearwise_count['Budget Amount']

budget_year.head(20)








```

##Plotting budget amount yearwise 

```python
budget_year.plot.bar(figsize=(9,9))

plt.xticks(rotation=0)

plt.ylabel('Budget in USD ($)')
```

##Monthwise 2008

```python
monthwise2008= budget[budget['BudgetYear']==2008]

monthwise2008=monthwise2008.groupby('BudgetPeriod').sum()

monthwise2008.head(30)


xlabels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


budget_2008_month=monthwise2008['Budget Amount']

plt.figure(figsize=(10,10))

plt.bar(xlabels,budget_2008_month)

plt.xlabel("Months")
plt.ylabel("Budget in USD ($)")

plt.title('2008 Monthly Budget')


```

##Monthwise 2009

```python
monthwise2009= budget[budget['BudgetYear']==2009]

monthwise2009=monthwise2009.groupby('BudgetPeriod').sum()

#monthwise2008.head(30)


#xlabels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


budget_2009_month=monthwise2009['Budget Amount']

plt.figure(figsize=(10,10))

plt.bar(xlabels,budget_2009_month)

plt.xlabel("Months")
plt.ylabel("Budget in USD ($)")

plt.title('2009 Monthly Budget')
```

```python
budget_monthly=(budget_2008_month+budget_2009_month)/2

plt.figure(figsize=(10,10))

plt.bar(xlabels,budget_monthly)

plt.xlabel("Months")
plt.ylabel("Budget in USD ($)")

plt.title('Average Monthly Budget')
```
